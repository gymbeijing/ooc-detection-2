```python
import matplotlib.pyplot as plt

from sklearn import datasets
from sklearn import manifold
import seaborn as sns
import json
import torch
import pandas as pd
import numpy as np
```

```python
emb = torch.load('../output/emb.pt')
z = torch.load('../output/z.pt')
topic = json.load(open('../output/topic.json', 'r'))
```

```python
tsne = manifold.TSNE(n_components=2, random_state=42)
```

```python
emb_proj = tsne.fit_transform(emb)
```

```python
z_proj = tsne.fit_transform(z)
```

```python
emb_df = pd.DataFrame(columns=['x1', 'x2', 'Topic'],
                       data=np.column_stack((emb_proj, 
                                            topic)))
```

```python
abbrev_dict = {"climate": "Cl", "covid": "Cv", "military": "M"}
emb_df.loc[:, 'x1'] = emb_df.x1.astype(float)
emb_df.loc[:, 'x2'] = emb_df.x2.astype(float)
emb_df['Topic'] = emb_df['Topic'].apply(lambda t: abbrev_dict[t.split('_')[0]])
```

```python
sns.set_theme(font_scale=1.5, style="white")
```

```python
color_scheme = ['lightblue', 'darkgray', 'lightcoral']
fig = sns.scatterplot(
    x="x1", y="x2",
    hue="Topic",
    data=emb_df,
#     palette=["C0", "C1"],
    # palette=sns.color_palette("hls", 3),
    palette=color_scheme,
    alpha=0.3,
)

fig.set(xlabel=None)
fig.set(ylabel=None)
fig.set(title="Multimodal feature x")
fig.set(xticklabels=[])
fig.set(yticklabels=[])
fig.tick_params(bottom=False)
fig.tick_params(left=False)
```

```python
z_df = pd.DataFrame(columns=['x1', 'x2', 'Topic'],
                       data=np.column_stack((z_proj, 
                                            topic)))
```

```python
z_df.loc[:, 'x1'] = z_df.x1.astype(float)
z_df.loc[:, 'x2'] = z_df.x2.astype(float)
z_df['Topic'] = z_df['Topic'].apply(lambda t: abbrev_dict[t.split('_')[0]])
```

```python
color_scheme = ['lightblue', 'darkgray', 'lightcoral']
fig = sns.scatterplot(
    x="x1", y="x2",
    hue="Topic",
    data=z_df,
#     palette=["C0", "C1"],
    # palette=sns.color_palette("hls", 3),
    palette = color_scheme,
    alpha=0.3,
)
# fig.set(title="T-SNE visualization of the latent variable z generated from $MMVAE_{deeper layers}$")
fig.set(xlabel=None)
fig.set(ylabel=None)
fig.set(title="Learned domain-invariant feature z")
fig.set(xticklabels=[])
fig.set(yticklabels=[])
fig.tick_params(bottom=False)
fig.tick_params(left=False)
```
