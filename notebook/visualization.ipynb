```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# da_settings = ("M", "Cl", "Cv", "B", "G", "U", "W")
da_settings = ("0.5", "1", "2", "5")
# acc = {
#     '0.5': (78.30, 80.18, 80.14, 71.37, 75.89, 79.98, 77.36),
#     '1': (79.26, 80.80, 80.06, 71.52, 75.70, 80.77, 77.17),
#     '2': (78.65, 80.80, 80.07, 71.48, 75.82, 79.96, 77.28),
#     '5': (78.89, 80.64, 79.74, 71.35, 75.88, 79.97, 77.38),
# }
acc = {
    'M': (78.30, 79.26, 78.65, 78.89),
    'Cl': (80.18, 80.80, 80.80, 80.64),
    'Cv': (80.14, 80.06, 80.07, 79.74),
}

x = np.arange(len(da_settings))  # the label locations
width = 0.2  # the width of the bars
multiplier = 0

fig, ax = plt.subplots()
color_scheme = ['lightcoral', 'lightblue', 'darkgray']

i = 0
for attribute, measurement in acc.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute, color=color_scheme[i])
    # ax.bar_label(rects, padding=3, fontsize=15)
    multiplier += 1
    i += 1

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Acc %', fontsize=30)
ax.set_xlabel('$\lambda_{MMD}$', fontsize=30)
ax.set_title('Twitter-COMMs', fontsize=30)
ax.set_xticks(x + width, da_settings, fontsize=30)
ax.legend(loc='upper left', ncol=3, fontsize=20)
ax.set_ylim(75, 82)

plt.rcParams["figure.figsize"] = [8, 6]
plt.rcParams["axes.linewidth"] = 2
plt.yticks(fontsize=30)
plt.show()
```
